```python
with open("data.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(text[:500])

data = text.split('\n')
data[:10]
```

```python
#giving index to every word (tokenize)
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data)
```

```python
tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(data)
sequences= [seq for seq in sequences if len(seq)>0]
sequences
```

```python
#making dataset
input_sequences = []
for sequence in sequences:
  # print(len(sequence))

  for i in range(1,len(sequence)):
      input_sequences.append(sequence[:i+1])

```

```python
input_sequences
```

```python
#padding, finding max length
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
#padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding="pre")
padded_input_sequences

#pre padding cuz itll be recent better for LSTM to remember
```

```python
#defining X and Y
X = padded_input_sequences[:,:-1] #remove last 1 from each
X
```

```python
Y = padded_input_sequences[:,-1] #last 1 from each
Y
```

```python
print(X.shape)
```

```python
Y.shape
```

```python
#total words
len(tokenizer.word_index)+1
#why +1 cuz our words start from index 1 but OHE start from index 0
#tokenizer starts from 1 to 107
```

```python
tokenizer.word_index
```

```python
#one hot encoding the output (considering this a multi classification problem)
from tensorflow.keras.utils import to_categorical
vocab_size = len(tokenizer.word_index) +1
print(vocab_size) # 8932
y = to_categorical(Y, num_classes=vocab_size)
```

```python
y.shape
```

```python
len(X[1])
```

```python
#building model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(vocab_size,150,input_length=19))
model.add(LSTM(50))
model.add(Dense(vocab_size, activation="softmax"))
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(X,y, epochs=100)
```

```python
# Training: text_on_sequence converts already indexed words into their original numeric sequence as in the sentence
# Prediction: converts new words into their original training indices (OOV words → 0) for model input
```

```python
import numpy as np
text = input("Enter the word: ")
num_words = int(input("Enter the Number of words you want: "))


for i in range(num_words):

  #tokenize
  test_word = tokenizer.texts_to_sequences([text])[0]
  #padd
  pa = pad_sequences([test_word], maxlen=17, padding="pre")


  #predict
  outout = model.predict(pa)
  outind = np.argmax(outout)
  outind

  output = tokenizer.index_word[outind]

  #next 5 words
  text = text + " " + output
print(text)
```
